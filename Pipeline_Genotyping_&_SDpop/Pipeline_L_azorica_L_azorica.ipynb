### 3. Script Transcriptome_sort/index/flagstat

Transcriptome_sort_index_flag.sh

```python
#!/bin/bash
#SBATCH --job-name=Transcriptome_sort/index/flagstat
#SBATCH -p unlimitq 
#SBATCH --time=7-00:00:00
#SBATCH --mail-user=tristan.bertrand@umontpellier.fr
#SBATCH --mail-type=ALL
#SBATCH --mem-per-cpu=8G
#SBATCH --cpus-per-task=12

#############################
# 1. MODULES
#############################

module load bioinfo/samtools/1.23

#############################
# 2. FILES 
#############################

SAMPLE="/home/tbertrand/work/L_azorica/L_azorica_L_azorica/MAPPING/INPUT/Samples.tsv"
MAPDIR="/home/tbertrand/work/L_azorica/L_azorica_L_azorica/MAPPING"

#############################
# 3. RUN Transcriptome sort/index/flagstat
#############################

# Récupère la 3e colonne (identifiants) et traite chaque échantillon
for IND in $(awk '{print $3}' "$SAMPLE"); do
    IN="${MAPDIR}/${IND}Aligned.toTranscriptome.out.bam"
    OUT_SORT="${MAPDIR}/${IND}Aligned.toTranscriptome.sorted.out.bam"
    OUT_FLAG="${MAPDIR}/${IND}Aligned.toTranscriptome.sorted.flagstat"

    samtools sort "$IN" -m 20000000000 -o "$OUT_SORT"
    samtools index "$OUT_SORT"
    samtools flagstat "$OUT_SORT" > "$OUT_FLAG"

done
```

### 4. Script Transform

```python
#!/bin/bash
#SBATCH --job-name=Transform
#SBATCH -p unlimitq 
#SBATCH --time=7-00:00:00
#SBATCH --mail-user=tristan.bertrand@umontpellier.fr
#SBATCH --mail-type=ALL
#SBATCH --mem-per-cpu=8G
#SBATCH --cpus-per-task=4
#SBATCH --array=1-32%8

module load bioinfo/samtools/1.23

SAMPLE="/home/tbertrand/work/L_azorica/L_azorica_L_azorica/MAPPING/INPUT/Samples.tsv"
MAPDIR="/home/tbertrand/work/L_azorica/L_azorica_L_azorica/MAPPING"

IND=$(awk -v line=$SLURM_ARRAY_TASK_ID 'NR==line {print $3}' "$SAMPLE")

echo "Processing $IND (task $SLURM_ARRAY_TASK_ID)"

BAM_TRANSCRIPTOME_SORTED="${MAPDIR}/${IND}Aligned.toTranscriptome.sorted.out.bam"
BAM_COORD="${MAPDIR}/${IND}Aligned.sortedByCoord.out.bam"

TRANSCRIPTOME_READ="${MAPDIR}/${IND}_transcriptome_reads.txt"
OUT_BAM="${MAPDIR}/${IND}_TranscriptomeGenomePos.bam"
OUT_RG_BAM="${MAPDIR}/${IND}_TranscriptomeGenomePos.rg.bam"
OUT_SORTED_RG_BAM="${MAPDIR}/${IND}_TranscriptomeGenomePos.sorted.rg.bam"

# extract read names
samtools view "$BAM_TRANSCRIPTOME_SORTED" | cut -f1 > "$TRANSCRIPTOME_READ"

# extract those reads from genomic bam
samtools view -N "$TRANSCRIPTOME_READ" -bh "$BAM_COORD" > "$OUT_BAM"

# Add ID in the BAM HEADER for multisample genotyping 
samtools addreplacerg \
    -r ID:${IND} \
    -r SM:${IND} \
    -r PL:ILLUMINA \
    -r LB:${IND} \
    -r PU:${IND} \
    -o "$OUT_RG_BAM" \
    "$OUT_BAM"

# sort + index
samtools sort -@ 4 -o "$OUT_SORTED_RG_BAM" "$OUT_RG_BAM"
samtools index "$OUT_SORTED_RG_BAM"

rm "$OUT_BAM"
rm "$OUT_RG_BAM"

done
```

## 2. SDPOP 

#### snakefile_SDpop

```python
# L objectif de ce script est prendre en INNPUT un fichier VCF brut (freebayes), de le formater pour SDpop et run SDpop

#######################################
# 1. CONFIG
#######################################

configfile: "config.yaml"

WORKDIR_SDPOP = config["WORKDIR_SDPOP"]
VCF           = config["VCF"]
ref_genome    = config["ref_genome"]
EXONS_BED     = config["EXONS_BED"]

VCF_NORM       = config["VCF_NORM"]
VCF_NORM_FILT  = config["VCF_NORM_FILT"]
VCF_NORM_FILT_EXONS_RECODE = config["VCF_NORM_FILT_EXONS_RECODE"]
CNT_FILE       = config["CNT_FILE"]

SDPOP_NOSEX_OUT = config["SDPOP_NOSEX_OUT"]
SDPOP_XY_OUT    = config["SDPOP_XY_OUT"]
SDPOP_ZW_OUT    = config["SDPOP_ZW_OUT"]

NOSEX_LOG = config["NOSEX_LOG"]
XY_LOG    = config["XY_LOG"]
ZW_LOG    = config["ZW_LOG"]


###########################################
# CIBLES FINALES 
###########################################

rule all:
    input:
        VCF_NORM,
        VCF_NORM_FILT,
        VCF_NORM_FILT_EXONS_RECODE,
        CNT_FILE,
        SDPOP_NOSEX_OUT,
        SDPOP_XY_OUT,
        SDPOP_ZW_OUT,
        NOSEX_LOG,
        XY_LOG,
        ZW_LOG


###############################################
# 3. RUN BCFTOOLS norm 
###############################################

rule bcftools_norm:
    input:
        vcf = VCF,
        ref = ref_genome
    output:
        VCF_NORM
    threads: 6
    resources:
        mem_mb=48000,
        slots=1,
        time="1-00:00:00"
    shell:
        """
        module load bioinfo/Bcftools/1.9

        bcftools norm --threads {threads} {input.vcf} -f {input.ref} -m +any -O v -o {output}
        """


##################################################
# 4. RUN FILTER VCFTOOLS 
##################################################

rule vcftools_filter:
    input:
        VCF_NORM
    output:
        VCF_NORM_FILT
    resources:
        mem_mb=8000,
        slots=1,
        time="1-00:00:00"
    shell:
        """
        module load bioinfo/samtools/1.14
        module load bioinfo/VCFtools/0.1.16

        vcftools --vcf {input} --minDP 7 --recode --out {output}
        mv {output}.recode.vcf {output}
        """


#########################################################
# 5. CUT EXONS WINDOWS 
#########################################################

rule exons_intersect:
    input:
        vcf = VCF_NORM_FILT,
        bed = EXONS_BED
    output:
        VCF_NORM_FILT_EXONS_RECODE
    resources:
        mem_mb=8000,
        slots=1,
        time="1-00:00:00"
    shell:
        """

        module load devel/python/Python-3.11.1
        module load bioinfo/bedtools/2.30.0

        # extract the header line (CHROM POS ...) of the vcf
        grep -m 1 "CHROM" "{input.vcf}" > "{output}"

        # create an intersect of the vcf and the bed file
        bedtools intersect -a "{input.vcf}" -b "{input.bed}" -wa -wb | \
        awk '{{
            ## We want to retrieve here the last column which corresponds to the name of the gene
            gene_id = $NF
            sub(/-.*/, "", gene_id)

            # print gene_id as the first column 
            printf "%s\t", gene_id

            # Print all the VCF column except the first (#CHROM) and the last 4 (BED column)
            for(i=2;i<=NF-4;i++) printf "%s\t", $i

            print ""
        }}' | sort -k1,1V -k2,2n >> "{output}"
        """


#############################################
# 6. RUN POPSUM 
#############################################

rule popsum_counts:
    input:
        VCF_NORM_FILT_EXONS_RECODE
    output:
        CNT_FILE
    resources:
        mem_mb=8000,
        slots=1,
        time="1-00:00:00"
    shell:
        """
        module load bioinfo/SDpop/f112148

        INDIV_LIST_FEMALES=$(grep "^#CHROM" "{input}" | cut -f10- | tr '\t' '\n' | grep '^LaF' | tr '\n' ',')
        INDIV_LIST_MALES=$(grep "^#CHROM" "{input}" | cut -f10- | tr '\t' '\n' | grep '^LaM' | tr '\n' ',')

        popsum "{input}" "{output}" v "$INDIV_LIST_FEMALES" "$INDIV_LIST_MALES"
        """


#############################################
# 7. RUN SDPOP 
#############################################

rule sdpop_nosex:
    input:
        CNT_FILE
    output:
        SDPOP_NOSEX_OUT
    log:
        NOSEX_LOG
    resources:
        mem_mb=16000,
        slots=1,
        time="1-00:00:00"
    shell:
        """
        module load bioinfo/SDpop/f112148
        sdpop "{input}" "{output}" e n 1 1 s > "{log}"
        """

rule sdpop_xy:
    input:
        CNT_FILE
    output:
        SDPOP_XY_OUT
    log:
        XY_LOG
    resources:
        mem_mb=16000,
        slots=1,
        time="1-00:00:00"
    shell:
        """
        module load bioinfo/SDpop/f112148
        sdpop "{input}" "{output}" e x 1 1 s > "{log}"
        """

rule sdpop_zw:
    input:
        CNT_FILE
    output:
        SDPOP_ZW_OUT
    log:
        ZW_LOG
    resources:
        mem_mb=16000,
        slots=1,
        time="1-00:00:00"
    shell:
        """
        module load bioinfo/SDpop/f112148
        sdpop "{input}" "{output}" e z 1 1 s > "{log}"
        """
```
