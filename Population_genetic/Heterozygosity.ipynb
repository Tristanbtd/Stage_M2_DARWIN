```python
#!/bin/bash

#SBATCH --job-name=Annotation_braker3
#SBATCH -p workq
#SBATCH --time=1-00:00:00
#SBATCH --mail-user=tristan.bertrand@supagro.fr
#SBATCH --mail-type=ALL
#SBATCH --mem-per-cpu=8G
#SBATCH --cpus-per-task=4

#############################
# 1. MODULES
#############################

module purge
module load devel/python/Python-3.6.3
module load bioinfo/Bcftools/1.9
module load bioinfo/samtools/1.14
module load bioinfo/VCFtools/0.1.16

#############################
# 2. FILES 
#############################

VCF=/home/tbertrand/work/VCF/L_azorica_on_L_nobilis/L_azorica_Lnobilis_VCFALLSITES_C2_REPMONO_NBESTALLELES2_norm.recode.vcf
OUTPUTDIR=/home/tbertrand/work/VCF/L_azorica_on_L_nobilis/stats
BASENAME=L_azorica_Lnobilis_VCFALLSITES_C2_REPMONO_NBESTALLELES2_norm.recode

#############################
# 3. RUN 
#############################

# Compression + indexation
bgzip -c $VCF > ${OUTPUTDIR}/${BASENAME}.vcf.gz
tabix -p vcf ${OUTPUTDIR}/${BASENAME}.vcf.gz

# Utilisation du vcf.gz pour les stats
VCFGZ=${OUTPUTDIR}/${BASENAME}.vcf.gz

# Global stats 
bcftools stats $VCFGZ > ${OUTPUTDIR}/${BASENAME}.global.stats

# Depth per individual 
vcftools --gzvcf $VCFGZ --depth --out ${OUTPUTDIR}/${BASENAME}.depth_indiv

# Depth per site 
vcftools --gzvcf $VCFGZ --site-mean-depth --out ${OUTPUTDIR}/${BASENAME}.depth_site

# Missingness
vcftools --gzvcf $VCFGZ --missing-indv --out ${OUTPUTDIR}/${BASENAME}.missing_indiv
vcftools --gzvcf $VCFGZ --missing-site --out ${OUTPUTDIR}/${BASENAME}.missing_site

# Heterozygosity
vcftools --gzvcf $VCFGZ --het --out ${OUTPUTDIR}/${BASENAME}.het

# Relatedness2
vcftools --gzvcf $VCFGZ --relatedness2 --out ${OUTPUTDIR}/${BASENAME}.relatedness2

# Site quality
vcftools --gzvcf $VCFGZ --site-quality --out ${OUTPUTDIR}/${BASENAME}.site_quality

# SNP density (fenêtres 100 kb)
vcftools --gzvcf $VCFGZ --SNPdensity 100000 --out ${OUTPUTDIR}/${BASENAME}.snp_density

# Singletons
vcftools --gzvcf $VCFGZ --singletons --out ${OUTPUTDIR}/${BASENAME}.singletons

```

```python
#!/usr/bin/env bash


# --- Entrée ---
VCF="$1"     # freebayes (g)VCF
OUT="filtered.vcf.gz"

# --- Paramètres filtres (ajuste si voulu) ---
MIN_DP=10
MAX_DP=200
MIN_GQ=20
MIN_QUAL=30
AB_MIN=0.25
AB_MAX=0.75

echo "[INFO] Normalisation + split multialléliques"
bcftools norm -m -both -Ou "$VCF" \
| bcftools view -Ou \
> tmp.norm.vcf

echo "[INFO] Filtre par génotype DP/GQ -> missing"
bcftools +setGT tmp.norm.vcf -Ou -- \
    -n . -t q \
    -e "FMT/DP<${MIN_DP} || FMT/DP>${MAX_DP} || FMT/GQ<${MIN_GQ}" \
> tmp.genofilt.vcf

echo "[INFO] Filtre AB pour hétérozygotes déséquilibrés"
bcftools +setGT tmp.genofilt.vcf -Ou -- \
    -n . -t q \
    -e 'GT="het" && (FMT/AO+FMT/RO)>0 && (FMT/AO)/(FMT/AO+FMT/RO)<'${AB_MIN}' || (FMT/AO)/(FMT/AO+FMT/RO)>'${AB_MAX}'' \
> tmp.ab.vcf

echo "[INFO] Filtre par site (QUAL), invariants conservés"
bcftools filter -e '(ALT!="." && QUAL<'${MIN_QUAL}')' tmp.ab.vcf -Ou \
> tmp.qual.vcf

echo "[INFO] Compression + index"
bgzip -c tmp.qual.vcf > "$OUT"
tabix -p vcf "$OUT"

rm tmp.*.vcf.*
rm tmp.*.vcf 2>/dev/null || true

echo "[OK] Résultat : $OUT"
```

```python
#!/bin/bash
#SBATCH --job-name=FILTER_VCF
#SBATCH -p workq
#SBATCH --time=1-00:00:00
#SBATCH --mem=12G
#SBATCH --cpus-per-task=1
#SBATCH --mail-type=END,FAIL


#############################
# MODULES
#############################
module purge
module load bioinfo/Bcftools/1.9
module load bioinfo/VCFtools/0.1.16

#############################
# INPUT / PARAMETERS
#############################
VCF="/home/tbertrand/work/L_azorica/L_azorica_L_azorica/GENOTYPING/TEST_FILTERS/L_azorica_Lazorica_VCFALLSITES_C2_REPMONO_NBESTALLELES2.vcf.gz"
OUT="filtered.vcf.gz"

MIN_DP=10
MIN_GQ=30
MIN_QUAL=30

#############################
# FILTERING PIPELINE
#############################

echo "[INFO] Filtering VCF..."

bcftools norm -m -both -Ou "$VCF" | \
bcftools +setGT -Ou -- \
    -n . -t q \
    -e "FMT/DP<${MIN_DP} || FMT/GQ<${MIN_GQ}" | \
bcftools filter -Ou \
    -e "(ALT != \".\" && QUAL < ${MIN_QUAL})" | \
bcftools view -Oz -o "$OUT"

tabix -p vcf "$OUT"


vcftools --remove-indels --minDP 10 --max-missing 0.9 --min-alleles 2 --max-alleles 2 --maf 0.05 --QUAL 30

#############################
# BASIC STATS
#############################

echo "[INFO] Generating stats..."

bcftools stats "$OUT" > filtered.global.stats
vcftools --gzvcf "$OUT" --missing-indv --out filtered.missing_indiv
vcftools --gzvcf "$OUT" --het --out filtered.het

echo "[INFO] Done."
```

```python
#!/bin/bash
#SBATCH --job-name=FILTER_VCFTOOLS
#SBATCH -p workq
#SBATCH --time=1-00:00:00
#SBATCH --mem=12G
#SBATCH --cpus-per-task=1
#SBATCH --mail-type=END,FAIL

#############################
# MODULES
#############################
module purge
module load bioinfo/Bcftools/1.9
module load bioinfo/VCFtools/0.1.16

#############################
# INPUT / PARAMETERS
#############################
VCF="/home/tbertrand/work/L_azorica/L_azorica_L_azorica/GENOTYPING/TEST_FILTERS/L_azorica_Lazorica_VCFALLSITES_C2_REPMONO_NBESTALLELES2.vcf.gz"
OUT="filtered"

#############################
# FILTERING PIPELINE
#############################

vcftools \
  --gzvcf "$VCF" \
  --remove-indels \
  --minDP 10 \
  --max-missing 0.9 \
  --min-alleles 2 \
  --max-alleles 2 \
  --maf 0.05 \
  --minQ 30 \
  --recode \
  --recode-INFO-all \
  --out "$OUT"

# GZIP the output VCF
bgzip -f ${OUT}.recode.vcf
OUTPUT_VCF="${OUT}.recode.vcf.gz"

#############################
# BASIC STATS
#############################

echo "[INFO] Generating stats..."

bcftools stats "$OUTPUT_VCF" > filtered.global.stats
vcftools --gzvcf "$OUTPUT_VCF" --missing-indv --out filtered.missing_indiv
vcftools --gzvcf "$OUTPUT_VCF" --het --out filtered.het

echo "[INFO] Done."
```
