## Extract SNP diagnostique from *L. nobilis* (POR) population

```python
cd /Users/cibio/Desktop/Laurus/SDPOP/Sexing_correction/SNP_diagnostic_L_nobilis_POR


#On garde uniquement les SNP avec noprior_xy=1 et on garde la ligne crespondnat au gene

awk '
# Nouveau bloc (ex: >g1, >g2, etc.)
# Si on avait déjà un bloc précédent et qu’il était "valide" on l’imprime avant de passer au suivant
/^>/{
    if(block && keep) printf "%s", block
    block=$0"\n"; keep=0; next
}
# On conserve les lignes du bloc seulement si la condition est remplie ($(NF-1)=noprior_xy = 1)
# ici proba posterieur SNP XY = 0.999
{
    if ($(NF-1) >= 1) {
        block = block $0 "\n"
        keep = 1
    }
}
# À la fin, on n’imprime le dernier bloc que s’il est valide
END{
    if(block && keep) printf "%s", block
}
' SDPOP_XY_L_nobilis_POR_Lazorica_norm_filtered_EXONS.recode.out > SDPOP_XY_L_nobilis_POR_Lazorica_norm_filtered_EXONS.recode_SNP_XY_1.out


#Garder les Position et ajouter le nom du CHR6. Attention a bien garder que des sites du CHR6

awk '
/^>/ {next}
{
    print "OZ345845.1\t" $1
}
' SDPOP_XY_L_nobilis_Lazorica_norm_filtered_EXONS.recode_SNP_XY_1.out > SNP_diagnostiques_L_nobilis_POR.txt


#L_nobilis_all
bcftools view -R SNP_diagnostiques_L_nobilis_POR.txt /Users/cibio/Desktop/Laurus/SDPOP/L_nobilis_all_L_azorica/L_nobilis_all_Lazorica_VCFALLSITES_C2_REPMONO_NBESTALLELES2.vcf.gz  > L_nobilis_all_diagnostic_snp_L_nobilis_POR.vcf

#L_novocanarienis
bcftools view -R SNP_diagnostiques_L_nobilis_POR.txt /Users/cibio/Desktop/Laurus/SDPOP/L_novocanariensis_L_azorica/SDPOP/L_novocanariensis_Lazorica_VCFALLSITES_C2_REPMONO_NBESTALLELES2.vcf.gz  > L_novocanariensis_diagnostic_snp_L_nobilis_POR.vcf


#L_azorica
#bcftools view -R SNP_diagnostiques_L_nobilis_POR.txt /Users/cibio/Desktop/Laurus/SDPOP/L_nobilis_all_L_azorica/L_nobilis_all_Lazorica_VCFALLSITES_C2_REPMONO_NBESTALLELES2.vcf.gz  > L_azorica_diagnostic_snp_L_nobilis_POR.vcf

```
